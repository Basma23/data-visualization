```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
movies = pd.read_csv('./MoviesOnStreamingPlatforms_updated.csv')
movies.head(5)
```

```python
X = movies['Year'].values
y = movies['Runtime'].values
```

```python
sns.relplot(x = 'Year', y = 'Runtime', hue = movies.Runtime.tolist(), data = movies);
```

```python
year_sums = movies.groupby('Year', as_index=False).sum()
year_mean = movies.groupby('Year', as_index=False).mean()
```

```python
year_mean
```

```python
sns.relplot(x = 'Year', y = 'Runtime', data = year_sums);
```

```python
sns.relplot(x = 'Year', y = 'Runtime', data = year_mean);
```

```python
sns.lmplot(x = "Year", y = "Runtime", data = year_sums);
```
