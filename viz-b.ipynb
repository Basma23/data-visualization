```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set()
```

```python
df = pd.read_csv('./master.csv')
df.head(5)
```

```python
sns.relplot(x = 'year', y = 'population', hue = df.suicides_no.tolist(), data = df);
```

```python
year_sums = df.groupby('year', as_index=False).sum()
year_mean = df.groupby('year', as_index=False).mean()
```

```python
year_sums
```

```python
year_mean
```

```python
sns.relplot(x = 'year', y = 'population', data = year_sums);
```

```python
sns.relplot(x = 'year', y = 'population', data = year_mean);
```

```python
sns.lmplot(x = 'year', y = 'population', data = year_sums);
```

```python
sns.lmplot(x = 'year', y = 'population', data = year_mean);
```

```python
graph = sns.barplot( x = "suicides_no", y = "population", data = df)
```

```python
sns.relplot(x = 'suicides_no', y = 'population', hue = df.year.tolist(), data = df);
```

```python
graph = sns.barplot( x = "suicides_no", y = "population", data = df)
```

```python
graph = sns.barplot( x = "suicides_no", y = "population", data = df, ci = None)
```

```python
import matplotlib.pyplot as plt
X = df['suicides_no'].values
height = X[:30]
bars = df['suicides_no'].head(30)
y_pos = np.arange(len(bars))
 
plt.bar(y_pos, height)
 
plt.xticks(y_pos, bars, rotation=90)
plt.show()
```
